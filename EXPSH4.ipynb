```python
origem = '/content/drive/My Drive/data/COMEXSTAT'
```

```python
arquivo_1 = origem + '/' + 'EXP_2021_MUN.csv'
arquivo_2 = origem + '/' + 'EXP_2022_MUN.csv'
arquivo_3 = origem + '/' + 'EXP_2023_MUN.csv'
arquivo_4 = origem + '/' + 'EXP_2024_MUN.csv'
arquivo_5 = origem + '/' + 'EXP_2025_MUN.csv'
arquivo_6 = origem + '/' + 'EXP_2026_MUN.csv'
sh4 = origem + '/' + 'NCM_SH.csv'
pais = origem + '/' + 'PAIS.csv'
mun = origem + '/' + 'UF_MUN.csv'
```

```python
exp21sh = pd.read_csv(arquivo_1,low_memory=False,sep= ';',encoding='UTF-8')
exp22sh = pd.read_csv(arquivo_2,low_memory=False,sep= ';',encoding='UTF-8')
exp23sh = pd.read_csv(arquivo_3,low_memory=False,sep= ';',encoding='UTF-8')
exp24sh = pd.read_csv(arquivo_4,low_memory=False,sep= ';',encoding='UTF-8')
exp25sh = pd.read_csv(arquivo_5,low_memory=False,sep= ';',encoding='UTF-8')
exp26sh = pd.read_csv(arquivo_6,low_memory=False,sep= ';',encoding='UTF-8')
exppais = pd.read_csv(pais,low_memory=False,sep= ';',encoding='latin1')
expmun = pd.read_csv(mun,low_memory=False,sep= ';',encoding='latin1')
expsh4 = pd.read_csv(sh4,low_memory=False,sep= ';',encoding='latin1')
```

```python
exp_final_sh4 = pd.concat([exp21sh,exp22sh,exp23sh,exp24sh,exp25sh,exp26sh],ignore_index=True)
```

```python
exp_final_sh4 = exp_final_sh4[exp_final_sh4['CO_MUN'].astype(str) == '3449904']
```

```python
exp_final_sh4 = exp_final_sh4.merge(exppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
exp_final_sh4 = exp_final_sh4.merge(expsh4[['SH4','NO_SH4_POR']],on='SH4',how='left')
```

```python
exp_final_sh4 = exp_final_sh4.merge(expmun[['CO_MUN','NO_MUN']],on='CO_MUN',how='left')
```
