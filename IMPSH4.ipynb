```python
origem = '/content/drive/My Drive/data/COMEXSTAT'
```

```python
arquivo_1 = origem + '/' + 'IMP_2021_MUN.csv'
arquivo_2 = origem + '/' + 'IMP_2022_MUN.csv'
arquivo_3 = origem + '/' + 'IMP_2023_MUN.csv'
arquivo_4 = origem + '/' + 'IMP_2024_MUN.csv'
arquivo_5 = origem + '/' + 'IMP_2025_MUN.csv'
arquivo_6 = origem + '/' + 'IMP_2026_MUN.csv'
sh4 = origem + '/' + 'NCM_SH.csv'
pais = origem + '/' + 'PAIS.csv'
mun = origem + '/' + 'UF_MUN.csv'
```

```python
imp21sh = pd.read_csv(arquivo_1,low_memory=False,sep= ';',encoding='UTF-8')
imp22sh = pd.read_csv(arquivo_2,low_memory=False,sep= ';',encoding='UTF-8')
imp23sh = pd.read_csv(arquivo_3,low_memory=False,sep= ';',encoding='UTF-8')
imp24sh = pd.read_csv(arquivo_4,low_memory=False,sep= ';',encoding='UTF-8')
imp25sh = pd.read_csv(arquivo_5,low_memory=False,sep= ';',encoding='UTF-8')
imp26sh = pd.read_csv(arquivo_6,low_memory=False,sep= ';',encoding='UTF-8')
imppais = pd.read_csv(pais,low_memory=False,sep= ';',encoding='latin1')
impmun = pd.read_csv(mun,low_memory=False,sep= ';',encoding='latin1')
impsh4 = pd.read_csv(sh4,low_memory=False,sep= ';',encoding='latin1')
```

```python
imp_final_sh4 = pd.concat([imp21sh,imp22sh,imp23sh,imp24sh,imp25sh,imp26sh],ignore_index=True)
```

```python
imp_final_sh4 = imp_final_sh4[imp_final_sh4['CO_MUN'].astype(str) == '3449904']
```

```python
imp_final_sh4 = imp_final_sh4.merge(imppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
imp_final_sh4 = imp_final_sh4.merge(impsh4[['SH4','NO_SH4_POR']],on='SH4',how='left')
```

```python
imp_final_sh4 = imp_final_sh4.merge(impmun[['CO_MUN','NO_MUN']],on='CO_MUN',how='left')
```
