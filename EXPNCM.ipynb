```python
origem = '/content/drive/My Drive/data/COMEXSTAT'
```

```python
arquivo_1 = origem + '/' + 'EXP_2021.csv'
arquivo_2 = origem + '/' + 'EXP_2022.csv'
arquivo_3 = origem + '/' + 'EXP_2023.csv'
arquivo_4 = origem + '/' + 'EXP_2024.csv'
arquivo_5 = origem + '/' + 'EXP_2025.csv'
arquivo_6 = origem + '/' + 'EXP_2026.csv'
ncm = origem + '/' + 'NCM.csv'
pais = origem + '/' + 'PAIS.csv'
vias = origem + '/' + 'VIA.csv'
urf = origem + '/' + 'URF.csv'
```

```python
exp21 = pd.read_csv(arquivo_1,low_memory=False,sep= ';',encoding='UTF-8')
exp22 = pd.read_csv(arquivo_2,low_memory=False,sep= ';',encoding='UTF-8')
exp23 = pd.read_csv(arquivo_3,low_memory=False,sep= ';',encoding='UTF-8')
exp24 = pd.read_csv(arquivo_4,low_memory=False,sep= ';',encoding='UTF-8')
exp25 = pd.read_csv(arquivo_5,low_memory=False,sep= ';',encoding='UTF-8')
exp26 = pd.read_csv(arquivo_6,low_memory=False,sep= ';',encoding='UTF-8')
exppais = pd.read_csv(pais,low_memory=False,sep= ';',encoding='latin1')
expvias = pd.read_csv(vias,low_memory=False,sep= ';',encoding='latin1')
expurf = pd.read_csv(urf,low_memory=False,sep= ';',encoding='latin1')
expncm = pd.read_csv(ncm, low_memory=False, sep=';', encoding='latin1')
```

```python
exp_final = pd.concat([exp21,exp22,exp23,exp24,exp25,exp26],ignore_index=True)
```

```python
exp_final = exp_final[exp_final['SG_UF_NCM'] == 'SP']
```

```python
exp_final = exp_final.merge(exppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
exp_final = exp_final.merge(expncm[['CO_NCM','NO_NCM_POR']],on='CO_NCM',how='left')
```

```python
exp_final = exp_final.merge(expurf[['CO_URF','NO_URF']],on='CO_URF',how='left')
```

```python
exp_final = exp_final.merge(expvias[['CO_VIA','NO_VIA']],on='CO_VIA',how='left')
```
