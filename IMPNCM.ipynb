```python
origem = '/content/drive/My Drive/data/COMEXSTAT'
```

```python
arquivo_1 = origem + '/' + 'IMP_2021.csv'
arquivo_2 = origem + '/' + 'IMP_2022.csv'
arquivo_3 = origem + '/' + 'IMP_2023.csv'
arquivo_4 = origem + '/' + 'IMP_2024.csv'
arquivo_5 = origem + '/' + 'IMP_2025.csv'
arquivo_6 = origem + '/' + 'IMP_2026.csv'
ncm = origem + '/' + 'NCM.csv'
pais = origem + '/' + 'PAIS.csv'
vias = origem + '/' + 'VIA.csv'
urf = origem + '/' + 'URF.csv'
```

```python
imp21 = pd.read_csv(arquivo_1,low_memory=False,sep= ';',encoding='UTF-8')
imp22 = pd.read_csv(arquivo_2,low_memory=False,sep= ';',encoding='UTF-8')
imp23 = pd.read_csv(arquivo_3,low_memory=False,sep= ';',encoding='UTF-8')
imp24 = pd.read_csv(arquivo_4,low_memory=False,sep= ';',encoding='UTF-8')
imp25 = pd.read_csv(arquivo_5,low_memory=False,sep= ';',encoding='UTF-8')
imp26 = pd.read_csv(arquivo_6,low_memory=False,sep= ';',encoding='UTF-8')
imppais = pd.read_csv(pais,low_memory=False,sep= ';',encoding='latin1')
impvias = pd.read_csv(vias,low_memory=False,sep= ';',encoding='latin1')
impurf = pd.read_csv(urf,low_memory=False,sep= ';',encoding='latin1')
impncm = pd.read_csv(ncm, low_memory=False, sep=';', encoding='latin1')
```

```python
imp_final = pd.concat([imp21,imp22,imp23,imp24,imp25,imp26],ignore_index=True)
```

```python
imp_final = imp_final[imp_final['SG_UF_NCM'] == 'SP']
```

```python
imp_final = imp_final.merge(imppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
imp_final = imp_final.merge(impncm[['CO_NCM','NO_NCM_POR']],on='CO_NCM',how='left')
```

```python
imp_final = imp_final.merge(impurf[['CO_URF','NO_URF']],on='CO_URF',how='left')
```

```python
imp_final = imp_final.merge(impvias[['CO_VIA','NO_VIA']],on='CO_VIA',how='left')
```
